## Import coco annotations 
Import annotations from the coco dataset, which are in coco json format. 

```python
import os 
import zipfile
import cv2

#Specify path to the coco.json file
path_to_annotations = "/content/drive/Shareddrives/Machine_Vision_Group_4/MinneApple_Datasets/test_data/detection/ground_truth.json"
#Specify the path to the images (if they are in a different folder than the annotations)
path_to_images = "/content/drive/Shareddrives/Machine_Vision_Group_4/MinneApple_Datasets/test_data/detection/images"

#Import the dataset into the pylable schema 
dataset = importer.ImportCoco(path_to_annotations, path_to_images=path_to_images, name="apples_COCO")
dataset.df.head(10)


```

```python
print(f"Number of images: {dataset.analyze.num_images}")
print(f"Number of classes: {dataset.analyze.num_classes}")
print(f"Classes:{dataset.analyze.classes}")
print(f"Class counts:\n{dataset.analyze.class_counts}")
print(f"Path to annotations:\n{dataset.path_to_annotations}")
```

# Export to Yolo v5
 

```python
dataset.path_to_annotations = "/content/drive/Shareddrives/Machine_Vision_Group_4/MinneApple_Datasets/test_data/detection"
dataset.export.ExportToYoloV5(output_path='/content/sample_data/textFiles', yaml_file='dataset.yaml',copy_images=False, use_splits=False, cat_id_index=None)
```
